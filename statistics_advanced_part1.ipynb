```python
1. What is a random variable in probability theory?
Ans-
A random variable is a numerical outcome of a random phenomenon or experiment. It assigns a real number to each possible outcome of an experiment. 
Random variables are essential because they help translate real-world random processes into mathematical terms that can be analyzed and computed. 
They are classified into two types: discrete random variables, which take specific, countable values, and continuous random variables, which take 
an infinite range of values within a given interval.
```

```python
2. What are the types of random variables?
Ans-
    The two main types of random variables are:
    Discrete Random Variables: These variables take on a finite or countable number of distinct values. Examples include the number of heads in coin tosses or the number of customers in a queue.
    Continuous Random Variables: These variables take on an infinite number of possible values within a range or interval. Examples include the height of individuals or the time required to complete a task.
```

```python
3. What is the difference between discrete and continuous distributions?
Ans-
    Discrete Distributions: These are associated with discrete random variables. The probabilities of all possible outcomes are listed individually, 
    often using a probability mass function (PMF). For instance, rolling a die has specific probabilities for outcomes 1 through 6.
    
    Continuous Distributions: These are associated with continuous random variables. Instead of individual probabilities, a probability density function 
    (PDF) is used, representing the density of probabilities over a range. Probabilities for specific intervals are calculated as areas under the PDF 
    curve.
```

```python
4.What are probability distribution functions (PDF)?
Ans-
    Probability distribution functions (PDF) describe the likelihood of a random variable taking on particular values.
    For discrete random variables, the PDF is a probability mass function (PMF), which assigns a specific probability to each value.
    For continuous random variables, the PDF represents the density of probability across a range of values. Unlike discrete PDFs, 
    the probability of an exact value is zero for continuous variables, so probabilities are determined over intervals.
```

```python
5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
Ans-
    PDF (Probability Distribution Function): Describes the likelihood of a random variable being exactly a certain value (for discrete) 
    or the density of probabilities (for continuous).
    
    CDF (Cumulative Distribution Function): Represents the probability that a random variable is less than or equal to a specific value. 
    The CDF is cumulative, summing probabilities for discrete variables or integrating the PDF for continuous variables. It always ranges 
    from 0 to 1 and provides a broader view of the distribution.
```

```python
6. What is a discrete uniform distribution?
Ans- 
    A discrete uniform distribution is one where all possible outcomes of a discrete random variable have the same probability. 
    For example, rolling a fair six-sided die results in an equal probability of ( \frac{1}{6} ) for each face. 
    This type of distribution is symmetric, and the probabilities do not vary with outcomes.
```

```python
7.  What are the key properties of a Bernoulli distribution?
Ans- 
    The Bernoulli distribution models a single trial with two outcomes: success (1) and failure (0).

   # Probability of Success: Denoted as ( p ).
   # Probability of Failure: Denoted as ( 1-p ).
   # Key Properties: It is used for binary events such as flipping a coin (heads or tails) or a yes/no decision. The mean of the distribution is ( p ), and the variance is ( p(1-p) ).
```

```python
8.What is the binomial distribution, and how is it used in probability?
Ans- 
    The binomial distribution extends the Bernoulli distribution to multiple trials. It describes the number of successes in a fixed number of independent
    trials of the same experiment, where each trial has the same probability of success (( p )). For example, it can model the probability of getting a 
    certain number of heads in 10 coin tosses. It is widely used in scenarios where events are repeated under identical conditions.
```

```python
9. What is the Poisson distribution, and where is it applied?
Ans- 
    The Poisson distribution describes the probability of a given number of events occurring in a fixed interval of time or space, 
    provided the events happen independently and at a constant average rate. It is commonly used in scenarios such as modeling the number of 
    customer arrivals at a store per hour or the number of emails received per day. The Poisson distribution is especially suited for rare events 
    over a large domain.
```

```python
10. What is a continuous uniform distribution?
Ans- 
    A continuous uniform distribution describes a random variable where all intervals of the same length within a given range are equally likely. 
    For example, if the waiting time for a bus is uniformly distributed between 0 and 30 minutes, each minute has the same probability density. 
    The PDF of a continuous uniform distribution is constant within the specified range and zero outside it.
```

```python
11. What are the characteristics of a normal distribution?
Ans- 
    The normal distribution, also known as the Gaussian distribution, is a symmetric, bell-shaped curve.

    Key Features:
        1.It is centered at the mean (( \mu )), with the spread determined by the standard deviation (( \sigma )).
        2.Most data points cluster around the mean, with probabilities decreasing as you move further away.
        3.It is fully described by its mean and standard deviation.
        4.The area under the curve represents probabilities, and the total area equals 1.
```

```python
12.What is the standard normal distribution, and why is it important?
Ans- 
    The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1.

    Importance: It provides a universal framework for comparing different normal distributions. By standardizing any normal random variable
    (converting to ( Z )-scores), we can use the standard normal table to calculate probabilities. This makes it a cornerstone of statistical methods.
```

```python
13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Ans- 
    The Central Limit Theorem states that the sampling distribution of the sample mean (or sum) of a sufficiently large number of independent 
    and identically distributed random variables approaches a normal distribution, regardless of the original distribution's shape.

    Importance: It enables statisticians to use normal distribution methods even when dealing with non-normal data, provided the sample size is large. 
    It forms the basis for hypothesis testing and confidence intervals.

```

```python
14.  How does the Central Limit Theorem relate to the normal distribution?
Ans- 
    The Central Limit Theorem establishes a direct connection to the normal distribution. It shows that as sample size increases, 
    the distribution of sample means becomes approximately normal, even if the underlying population distribution is not normal. 
    This relationship underpins many statistical analyses, enabling practitioners to apply normal distribution techniques to a wide range of 
    real-world problems.
```

```python
15. Application of Z Statistics in Hypothesis Testing
Ans- 
    Z-statistics are widely used in hypothesis testing to determine whether there is enough evidence to reject a null hypothesis. 
    They are applied when the population variance is known or the sample size is large (typically ( n > 30 )). The Z-test compares 
    the sample mean to the population mean and determines how many standard deviations the sample mean is away from the population mean. 
    This helps assess whether observed differences are due to chance or a significant effect. Common applications include testing population proportions,
    comparing means, and assessing the effectiveness of interventions or treatments.
```

```python
16. How to Calculate a Z-Score and Its Representation
Ans-
    A Z-score, also known as a standard score, represents the number of standard deviations a data point is away from the mean of a distribution. 
    It standardizes values from different scales, making them comparable. The formula for calculating a Z-score involves subtracting the mean from 
    the data point and dividing by the standard deviation.

    A Z-score provides a way to determine how unusual or typical a data point is in a given distribution. 
    Positive Z-scores indicate values above the mean, negative scores indicate values below the mean, and a Z-score of 0 means the value is exactly 
    at the mean.
```

```python
17. Point Estimates and Interval Estimates in Statistics
Ans- 
    A point estimate is a single value that serves as the best guess or approximation of a population parameter. For instance, the sample mean is 
    a point estimate of the population mean.

    An interval estimate provides a range of values within which the population parameter is likely to lie, incorporating the variability in the data. 
    Confidence intervals are a common example, offering a range based on a specified level of confidence (e.g., 95%). Point estimates give precision,
    while interval estimates account for uncertainty and provide a broader context.
```

```python
18. Significance of Confidence Intervals in Statistical Analysis
Ans- 
    Confidence intervals are crucial in statistical analysis because they quantify the uncertainty around a sample estimate. 
    Instead of relying solely on a single value (point estimate), confidence intervals give a range that is likely to contain the true population 
    parameter. This range is associated with a confidence level, such as 95%, meaning that if the experiment were repeated numerous times, 95% of the
    calculated intervals would contain the true parameter. Confidence intervals are particularly useful for communicating the precision and reliability of
    statistical estimates.
```

```python
19. Relationship Between a Z-Score and a Confidence Interval
Ans- 
    The Z-score plays a pivotal role in constructing confidence intervals for data that follows a normal distribution. 
    A Z-score corresponds to a specific percentile in the standard normal distribution, which determines the boundaries of the confidence interval.
    For example, for a 95% confidence interval, the Z-scores typically used are approximately ±1.96. These scores define the range within which the 
    population parameter is likely to fall with the desired level of confidence.
```

```python
20. Using Z-Scores to Compare Different Distributions
Ans- 
    Z-scores enable comparisons between different distributions by standardizing values. This standardization removes the influence of different 
    means and standard deviations, allowing values from different datasets or measurements to be compared on a common scale. For example, a Z-score 
    can help compare the test scores of students from different schools or scales by indicating how far a score is from its respective mean in terms of 
    standard deviations.
```

```python
21.  Assumptions for Applying the Central Limit Theorem
Ans- 
    The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as 
    the sample size increases, regardless of the population's original distribution. The assumptions for applying the CLT include:

    Independence: Samples should be independent of one another.
    Random Sampling: Data should be collected using random sampling methods.
    Sample Size: A sufficiently large sample size is required (typically ( n > 30 )).
    Finite Variance: The population should have a finite mean and variance.
```

```python
22. Concept of Expected Value in a Probability Distribution
Ans- 
    The expected value is the theoretical mean of a random variable in a probability distribution. It represents the long-term average outcome of a 
    random process if repeated infinitely. For discrete variables, it is calculated as the sum of each possible outcome multiplied by its probability.
    For continuous variables, it involves an integral. The expected value is a key concept in decision-making, providing a basis for evaluating the
    outcomes of uncertain events.
```

```python
23. Relation Between a Probability Distribution and Expected Outcome
Ans- 
    A probability distribution describes how the probabilities of different outcomes are distributed for a random variable. 
    The expected outcome, or expected value, is derived from this distribution. It acts as the weighted average of all possible outcomes, 
    with weights corresponding to their probabilities. The distribution shapes the expected value, as it accounts for both the likelihood and 
    magnitude of outcomes, offering insights into the variable's behavior over repeated trials.
```

```python
#Q :- 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
#Answer :-
def bernoulli_pdf(p, x):
    """
    Calculate the PDF of a Bernoulli distribution.
    p: Probability of success (0 <= p <= 1)
    x: Outcome (0 or 1)
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1.")
    return p if x == 1 else 1 - p

# Test the function
p = 0.6  # Probability of success
x = 1    # Outcome
print(f"PDF of Bernoulli({p}, {x}): {bernoulli_pdf(p, x)}")
```

```python
#Q :- 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
#Answer :-
import numpy as np
import matplotlib.pyplot as plt

# Parameters for discrete uniform distribution
low, high = 1, 6

# Generate random values
data = np.random.randint(low, high + 1, size=10000)

# Compute the CDF
values, counts = np.unique(data, return_counts=True)
cdf = np.cumsum(counts) / len(data)

# Plot the CDF
plt.step(values, cdf, where='mid', color='purple')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.grid()
plt.show()
```

```python
#Q :- 9. Write a Python function to calculate Z-scores from a dataset and plot them.
#Answer :-
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.
    """
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

# Generate random dataset
data = np.random.normal(50, 10, 1000)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot Z-scores histogram
plt.hist(z_scores, bins=50, density=True, alpha=0.7, color='teal')
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
#Q :- 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
#Answer :-
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_samples = 1000
sample_size = 30
iterations = 1000

# Generate a non-normal distribution (Exponential)
data = np.random.exponential(scale=2, size=n_samples)

# Apply CLT
means = [np.mean(np.random.choice(data, size=sample_size)) for _ in range(iterations)]

# Plot the results
plt.hist(means, bins=50, density=True, alpha=0.7, color='purple')
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
#Q :- 15. Implement hypothesis testing using Z-statistics for a sample dataset.
#Answer :-
from scipy.stats import norm

# Parameters
sample_mean = 51
population_mean = 50
std_dev = 10
sample_size = 30

# Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))

# Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#Q :- 16. Create a confidence interval for a dataset using Python and interpret the result.
#Answer :-
# Parameters
data = np.random.normal(50, 10, 100)
mean = np.mean(data)
std_error = np.std(data, ddof=1) / np.sqrt(len(data))

# 95% confidence interval
z_critical = norm.ppf(0.975)  # 95% CI
margin_of_error = z_critical * std_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print(f"Confidence Interval: {confidence_interval}")
```
